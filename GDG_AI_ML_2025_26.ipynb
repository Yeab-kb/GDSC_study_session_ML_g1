```python
import numpy as np

numbers = [1, 2, 3, 4, 5]

array = np.array(numbers)

mean_value = np.mean(array)
max_value = np.max(array)
sum_value = np.sum(array)

print("Mean:", mean_value)
print("Maximum Value:", max_value)
print("Sum:", sum_value)
```
